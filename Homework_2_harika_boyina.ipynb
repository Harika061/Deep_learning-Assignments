```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_datasets as tfds
```

```python
import tensorflow as tf
from tensorflow import keras

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize the pixel values to be between 0 and 1
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
class_names = ['Zero', 'One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine']
```

```python
plt.figure()
plt.imshow(x_train[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize = (10,10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap = plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

```python
# Define the model architecture
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Compile the model with appropriate loss and metrics
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model on the training data
model.fit(x_train, y_train, epochs=5)

# Evaluate the model on the training data
train_loss, train_acc = model.evaluate(x_train, y_train, verbose=2)
print(f"Training accuracy: {train_acc*100:.2f}%")

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")
```

```python
#Add another Dense layer of 128 nodes
model1 = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(128, activation='relu'),  # New layer with 128 nodes
    keras.layers.Dense(10, activation='softmax')
])

# Compile the model with appropriate loss and metrics
model1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model on the training data
model1.fit(x_train, y_train, epochs=5)

# Evaluate the model on the training data
train_loss, train_acc = model1.evaluate(x_train, y_train, verbose=2)
print(f"Training accuracy: {train_acc*100:.2f}%")

# Evaluate the model on the test data
test_loss, test_acc = model1.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")
```

```python
#Increase the current number of nodes in the layer to 256
model2 = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(256, activation='relu'),  # New layer with 256 nodes
    keras.layers.Dense(10, activation='softmax')
])

# Compile the model with appropriate loss and metrics
model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model on the training data
model2.fit(x_train, y_train, epochs=5)

# Evaluate the model on the training data
train_loss, train_acc = model2.evaluate(x_train, y_train, verbose=2)
print(f"Training accuracy: {train_acc*100:.2f}%")

# Evaluate the model on the test data
test_loss, test_acc = model2.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")
```

```python
#Experimenting with different optimizers, loss functions, dropout,  activation functions
#Dropout added-0.2
model3 = keras.models.Sequential([
        keras.layers.Flatten(input_shape=(28, 28)),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(10, activation='softmax')
    ])
# Compile the model with appropriate loss and metrics
model3.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# Train the model on the training data
model3.fit(x_train, y_train, epochs=5)

# Evaluate the model on the training data
train_loss, train_acc = model3.evaluate(x_train, y_train, verbose=2)
print(f"Training accuracy: {train_acc*100:.2f}%")

# Evaluate the model on the test data
test_loss, test_acc = model3.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")
```

```python
# Experiment with different optimizers
# activation func-relu
model4 = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape = (28, 28)), tf.keras.layers.Dense(256, activation='relu'), tf.keras.layers.Dense(10)])
model4.add(tf.keras.layers.Dense(256, activation='relu'))
model4.compile(optimizer = 'adam', loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
# Train the model on the training data
model4.fit(x_train, y_train, epochs=5)

# Evaluate the model on the training data
train_loss, train_acc = model4.evaluate(x_train, y_train, verbose=2)
print(f"Training accuracy: {train_acc*100:.2f}%")

# Evaluate the model on the test data
test_loss, test_acc = model4.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")

model5 = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape = (28, 28)), tf.keras.layers.Dense(256, activation='relu'), tf.keras.layers.Dense(10)])
model5.add(tf.keras.layers.Dense(256, activation='relu'))
model5.compile(optimizer = 'sgd', loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
# Train the model on the training data
model5.fit(x_train, y_train, epochs=5)

# Evaluate the model on the training data
train_loss, train_acc = model5.evaluate(x_train, y_train, verbose=2)
print(f"Training accuracy: {train_acc*100:.2f}%")

# Evaluate the model on the test data
test_loss, test_acc = model5.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")
```

```python
# Experiment with different loss functions

model6 = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape = (28, 28)), tf.keras.layers.Dense(256, activation='relu'), tf.keras.layers.Dense(10)])
model6.add(tf.keras.layers.Dense(256, activation='relu'))
model6.compile(optimizer = 'adam', loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
# Train the model on the training data
model6.fit(x_train, y_train, epochs=5)

# Evaluate the model on the training data
train_loss, train_acc = model6.evaluate(x_train, y_train, verbose=2)
print(f"Training accuracy: {train_acc*100:.2f}%")

# Evaluate the model on the test data
test_loss, test_acc = model6.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")
```

Here when I applied the modification to the model that is Added another Dense layer of 128 nodes, the train and test accuracy increased to 99.16% and 97.82%. With the other modification to the model that is Increase the current number of nodes in the layer to 256, , the train and test accuracy increased to 99.32% and 98.10%.
Also I have tried with with different optimizers, loss functions, dropout,  activation functions, but none of the model got a better train and test accuracy that the above modifications.
